```python
import pandas as pd
import numpy as np
```

```python
#Moving Average
#Rolling- Take a Moving Window over the data and apply calculation.
#Window - Size of Group
#Moving the group step by step
df = pd.DataFrame({"Months":["Jan","Feb","March","April"],
                   "Sales":[200,400,500,700]
                   })
df["MA"] =df["Sales"].rolling(window=3).mean()
df
```

```python
#Weighted Moving Average
window = df["Sales"].tail(3).values
print(window)



weight = np.arange(1,len(window)+1)
print(weight)

#Normalize Weight this is used
WA = weight/weight.sum()
print(WA)

# WAA = window/window.sum()
# print(WAA)
```

```python
#ForeCast

Forcast = np.sum(window*WA)
print(Forcast)
```

```python
#Arima
#Order = (p,d,q)
# p = auto regressive
# d = diffrence
# q = moving error


import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf

from stastsmodel.tsa.arima.model import ARIMA
plot_acf(df["Sales"])
plt.show()

plot_pacf(df["Sales"])
plt.show()
```

```python
#Arima

arima = ARIMA(df["Sales"]),(order = (0,2,0))
FA = arima.fit()
```

```python
forcast1 = FA.forecast(step =1)
```
