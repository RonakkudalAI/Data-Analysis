```python

# Central Limit Theorem Demo
import numpy as np
import matplotlib.pyplot as plt

# Original population (not normal)
#population = np.random.exponential(scale=2, size=10000)
population = np.random.uniform(0,10,10000)
# Take samples and compute means
sample_means = []
sample_size = 30   # n >= 30 (CLT condition)

for _ in range(1000):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))
    sample_std= np.std(sample)
# Plot original population
plt.figure()
plt.hist(population, bins=50)
plt.title("Original Population (Not Normal)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid()
plt.show()

# Plot sampling distribution (CLT result)
plt.figure()
plt.hist(sample_means, bins=50,density=True)
plt.title("Sampling Distribution (CLT - Normal)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```
