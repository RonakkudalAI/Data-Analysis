```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

df = pd.DataFrame({
    "study_hours": [2, 4, 6, 8, 10],
    "marks": [10, 20, 30, 40, 100,200]
})
```

```python
number_only = df.select_dtypes(include=["int64", "float64"])
```

```python
plt.figure(figsize=(10,6))
number_only.boxplot()
plt.title("With Outlier")
plt.show()
```

```python
# Remove Outlier
Q1 = number_only.quantile(0.25)
Q3 = number_only.quantile(0.75)

IQR = Q3-Q1

lower_range = Q1-1.5*IQR
upper_range = Q3+1.5*IQR
df_clean = df[~((number_only<lower_range) | (number_only>upper_range)).any(axis=1)]
print(df_clean)
```

```python
plt.figure(figsize=(10,6))
df_clean.boxplot()
plt.title("Without Outlier")
plt.show()
```
