```python
import pandas as pd
df = pd.read_csv("/content/superstore - superstore.csv")
df.info()
df.head()
```

```python
#Drop Column that is not important
df.drop("记录数",axis=1,inplace = True)
```

```python
#Statistical Summary-
df.describe()
```

```python
#Find Missing Value -:
df.isnull().sum() #isna()
# fillna for fill value
# for fill median use
```

```python
Category_sales = df.groupby("Category")["Sales"].sum()
Category_sales
```

```python
Category_profit = df.groupby("Category")["Profit"].sum()
Category_profit
```

```python
import matplotlib.pyplot as plt
Category_sales.plot(kind="bar")
plt.title("Category Vs Sales")
plt.xlabel("Category")
plt.ylabel("Sales")
# plt.grid(axis ="y")
for i,values in enumerate(Category_sales):
  plt.text(i,values,str(values),ha="center")
plt.show()
```

```python
from matplotlib import figure
plt.figure(figsize=(10,6))
plt.pie(Category_sales,labels=Category_sales.index,autopct="%1.1f%%",startangle=90)
plt.title("Category VS Sales")
plt.show()
```
