```python
#Quantile And IQR
Q1 = df.quantile(0.25,numeric_only=True)
Q3 = df.quantile(0.75,numeric_only=True)

IQR = Q3-Q1
print(IQR)
```

```python
#Outlier Detections
lower_range = Q1-1.5*IQR
upper_range = Q3+1.5*IQR
print(lower_range)

print(upper_range)
```

```python
colums = df.select_dtypes(["int64","float64"])
outlier = colums[(colums<lower_range)|(colums>upper_range)]
outlier
```

```python
outlier = [(colums<lower_range)|(colums>upper_range)]
outlier
```

```python
outlier = colums[((colums<lower_range)|(colums>upper_range)).any(axis=1)]
outlier
```

```python
outlier1 = colums[((colums>=lower_range)|(colums<=upper_range)).any(axis=1)]
outlier1
```

```python
outlier = colums[~((colums<lower_range)|(colums>upper_range)).any(axis=1)]
outlier.describe()
print(type(outlier))
```
