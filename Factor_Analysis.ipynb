```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Math":[85,78,92,70,88],
    "Physics":[80,75,89,65,84],
    "Chemistry":[82,77,90,68,86],
    "English":[60,65,58,70,62],
    "History":[55,60,52,68,57]
}
df = pd.DataFrame(data)
df
```

```python
arr = df.to_numpy()
print(arr)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(arr)
print(eigen_values)
```

```python
print(eigen_vectors)
```

```python
for i in range(len(eigen_values)):

    if eigen_values[i] > 1:
        print("Eigen Value:", eigen_values[i])
        print("Corresponding Eigen Vector:", eigen_vectors[:, i])
        print("----------------------------")
```

```python
from sklearn.decomposition import FactorAnalysis
```

```python
#Model Create
fa = FactorAnalysis(n_components=2)

# Fit + Transform
factors = fa.fit_transform(df)

print(factors)
```
