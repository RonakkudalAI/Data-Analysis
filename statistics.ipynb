```python
import numpy as np
```

```python
A = np.array([45,50,55,60,67,70])
print("25 percentile :",np.percentile(A,25))
print("25 percentile midpoint :",np.percentile(A,25,method="midpoint"))
print("25 percentile nearest :",np.percentile(A,25,method="nearest"))
print("50 percentile :",np.percentile(A,50))
print("75 percentile :",np.percentile(A,75))
```

```python
print(np.quantile(A,0.25))
print(np.quantile(A,0.5))
print(np.quantile(A,0.75))
```

```python
A1 = np.array([45,50,55,60,67,70,75])
print("25 percentile :",np.percentile(A1,25))
print("25 percentile midpoint :",np.percentile(A1,25,method="midpoint"))
print("25 percentile nearest :",np.percentile(A1,25,method="nearest"))
print("50 percentile :",np.percentile(A1,50))
print("75 percentile :",np.percentile(A1,75))
```

```python
print(np.quantile(A1,0.25))
print(np.quantile(A1,0.5))
print(np.quantile(A1,0.75))
```

```python
Q1,Q2,Q3 = np.percentile(A1,[25,50,75])
print(Q1,Q2,Q3)
```

```python
#IQR
print(Q3- Q1)
```

```python
A1 = np.array([45,50,55,60,67,70])
print("Variance",np.var(A1))
```

```python
A = np.array([45,50,55,60,67,70,75])
print("Variance :",np.var(A))
```

```python
#Sample variance
np.var(A1,ddof = 1)
```

```python
std_dev = np.sqrt(np.var(A))
print(std_dev)
```

```python
print(np.std(A))
```
